## Cell 2: Data Configuration — URLs and Directory Setup

**How Claude helped:** My original code defined three completely separate dictionaries — `CRASHES`, `PARTIES`, and `INJURED_WITNESSES` — each mapping a year to a URL. Claude pointed out (reinforcing what we discussed in class) that this is repetitive and harder to maintain. If I needed to add a new year or fix a URL, I'd have to update three different places. Claude refactored these into a single nested dictionary called `DATA_URLS`, where the outer key is the year (2021–2025) and the inner keys are the dataset types (`"crashes"`, `"parties"`, `"injured_witnesses"`). `YEARS` is now derived directly from `DATA_URLS` so the two can never get out of sync.

**What I learned:** The standard approach in data journalism notebooks is to keep imports in the very first cell and manage dependencies through a `requirements.txt` file rather than installing packages inline. It keeps the notebook cleaner and more professional.

```python
def download_if_missing(url: str, dest: Path) -> None:
    """Download file from URL if it doesn't exist locally or is empty."""
    if dest.exists() and dest.stat().st_size > 0:
        print(f"exists: {dest.name}")
        return
    
    tmp = dest.with_suffix(dest.suffix + ".part")
    
    try:
        urllib.request.urlretrieve(url, tmp)
        tmp.replace(dest)
    except Exception as e:
        if tmp.exists():
            tmp.unlink()
        raise
    finally:
        if tmp.exists():
            tmp.unlink(missing_ok=True)
def download_all_data() -> None:
    """Download all required data files."""
    
    for year in YEARS:
        print(f"\n--- Year {year} ---")
        download_if_missing(CRASHES[year], RAW_DIR / f"crashes_{year}.csv")
        download_if_missing(PARTIES[year], RAW_DIR / f"parties_{year}.csv")
        download_if_missing(INJURED_WITNESSES[year], RAW_DIR / f"injured_witnesses_{year}.csv")
    
# Run the download
download_all_data()
```

## Cell 3: Download Functions

**How Claude helped:** After restructuring `DATA_URLS` into a nested dictionary, the original `download_all_data()` function — which called `download_if_missing` three separate times per year for `CRASHES`, `PARTIES`, and `INJURED_WITNESSES` — was no longer aligned with the new structure. Claude updated the function to loop over `DATA_URLS.items()` and then loop over each year's inner dictionary, so adding a new year or a new dataset type only requires updating `DATA_URLS` — the download logic stays untouched.

**What I learned:** When you improve the organization of your data structures, the functions that use them should be updated to match. A function that iterates a nested dictionary with two loops is cleaner than one that repeats the same operation three times by hand. This reduces the chance of making a mistake if the structure changes later.

```python
def norm(column_name: str) -> str:
    """Normalize column names for consistent matching."""
    return "".join(ch for ch in str(column_name).strip().lower() if ch.isalnum())
def build_address(row: pd.Series) -> str:
    """Build a readable address from crash location components."""
    primary = str(row.get("PrimaryRoad", "")).strip()
    secondary = str(row.get("SecondaryRoad", "")).strip()
    sec_direction = str(row.get("SecondaryDirection", "")).strip()
    sec_distance = str(row.get("SecondaryDistance", "")).strip()
    if primary and secondary and secondary.lower() != "nan":
        return f"{primary} & {secondary}"
    if primary:
        if (
            sec_distance
            and sec_distance.lower() != "nan"
            and sec_direction
            and sec_direction.lower() != "nan"
        ):
            return f"{sec_direction} {sec_distance} from {primary}"
        return primary
    return ""
def load_crashes_data(year: int) -> pd.DataFrame:
    """Load and return crashes data for a given year."""
    crashes_path = RAW_DIR / f"crashes_{year}.csv"
    if not crashes_path.exists():
        raise FileNotFoundError(f"Crashes data for {year} not found: {crashes_path}")
    
    return pd.read_csv(crashes_path, low_memory=False)
def load_parties_data_chunked(year: int, chunk_size: int = 200_000):
    """Load parties data in chunks."""
    parties_path = RAW_DIR / f"parties_{year}.csv"
    if not parties_path.exists():
        raise FileNotFoundError(f"Parties data for {year} not found: {parties_path}")
    
    return pd.read_csv(
        parties_path,
        chunksize=chunk_size,
        engine="python",
        on_bad_lines="skip",
    )
print("Utility functions defined")
```

## Cell 4: Utility Functions

**How Claude helped:** Claude walked me through what each of these helper functions does so I could understand why they exist before the main analysis runs. `norm()` strips whitespace, lowercases, and removes non-alphanumeric characters from column names — this is necessary because the CA open data CSVs sometimes have inconsistent column naming (extra spaces, different casing) across years. `build_address()` constructs a readable street address from multiple location columns, handling cases where only a primary road exists or where a secondary road is also present. `load_crashes_data()` and `load_parties_data_chunked()` encapsulate file-reading logic so the main analysis function stays clean — loading parties in chunks is specifically needed because those files can be hundreds of thousands of rows.

**What I learned:** Breaking repetitive or complex operations into small named functions makes the main logic much easier to read and debug. The chunked loading pattern is an important technique for working with large files that might not fit in memory all at once.

```python
def run_analysis() -> Tuple[pd.DataFrame, pd.DataFrame, pd.DataFrame, pd.DataFrame]:
    """
    Run the complete SF crash analysis.
    
    Returns:
        Tuple of (age_frequency, nonblank_age_frequency, bike_counts, bike_locations)
    """
    # These lists will hold data that we collect year by year.
    all_non_driver_ages = []
    yearly_bike_counts = []
    bike_location_frames = []
    for year in YEARS:
        
        # Load crashes data
        crashes = load_crashes_data(year)
        crash_cols = {norm(c): c for c in crashes.columns}
        # Extract required column names
        col_collision = crash_cols.get("collisionid")
        col_city = crash_cols.get("cityname")
        col_primary = crash_cols.get("primaryroad")
        col_secondary = crash_cols.get("secondaryroad")
        col_secdir = crash_cols.get("secondarydirection")
        col_secdist = crash_cols.get("secondarydistance")
        col_lat = crash_cols.get("latitude")
        col_lon = crash_cols.get("longitude")
        col_dt = crash_cols.get("crashdatetime")
        if col_collision is None or col_city is None:
            raise ValueError(f"Required crash columns missing in {year} data")
        # --------- FILTER TO SAN FRANCISCO ONLY ---------
        # We only want to study crashes that occurred in San Francisco,
        # not across the entire state. So this line filters out other cities.
        sf_crashes = crashes[
            crashes[col_city].astype(str).str.strip().str.upper().eq("SAN FRANCISCO")
        ].copy()
        sf_crashes[col_collision] = sf_crashes[col_collision].astype(str).str.strip()
        # Record a set of all unique "collision IDs" for San Francisco crashes.
        # Will use these IDs later to link crashes to the people or parties involved.
        sf_collision_ids = set(sf_crashes[col_collision].dropna().tolist())
        
        print(f"  SF crashes: {len(sf_collision_ids):,}")
        # Process parties data in chunks
        chunk_iter = load_parties_data_chunked(year)
        # Temporary collections for this year's loop
        non_driver_age_rows = [] # Non-driver people and their ages
        bike_party_count = 0 # How many people were bicyclists
        bike_collision_ids = set() # The IDs of crashes that involved bicyclists
        sf_party_row_count = 0 # Total number of party rows belonging to SF crashes
        # For each chunk of party data:
        # Filter down to only those that match the SF collision IDs, and
        # classify individuals by their role (driver, pedestrian, bicyclist)
        for chunk_num, chunk in enumerate(chunk_iter, 1):
            if chunk_num % 5 == 0:
                print(f"    Processing chunk {chunk_num}...")

            party_cols = {norm(c): c for c in chunk.columns}
            p_col_collision = party_cols.get("collisionid")
            p_col_partytype = party_cols.get("partytype")
            p_col_age = party_cols.get("statedage")
            if p_col_collision is None or p_col_partytype is None:
                raise ValueError(f"Required party columns missing in {year} data")
            # Clean up the collision ID column for consistent matching
            chunk[p_col_collision] = chunk[p_col_collision].astype(str).str.strip()
            # Keep only parties (people/vehicles) involved in San Francisco crashes
            sf_parties = chunk[chunk[p_col_collision].isin(sf_collision_ids)].copy()
            if sf_parties.empty:
                continue
            sf_party_row_count += len(sf_parties)
            # Make all party types uppercase for consistent comparison.
            party_type_upper = (
                sf_parties[p_col_partytype].astype(str).str.strip().str.upper()
            )
            # We are interested in the role of people who were *not* the driver
            # (for example, passengers, pedestrians, or bicyclists).
            non_driver_mask = ~party_type_upper.eq("DRIVER")
            # Store ages of these non-driver participants (if the age data exists)
            if p_col_age is not None:
                ages = sf_parties.loc[non_driver_mask, [p_col_age]].copy()
                ages.columns = ["StatedAge"]
                non_driver_age_rows.append(ages)
            # Count how many people were bicyclists and which crashes those were.
            bike_mask = party_type_upper.str.contains("BICYCL", na=False)
            bike_party_count += int(bike_mask.sum())
            if bike_mask.any():
                bike_collision_ids.update(
                    sf_parties.loc[bike_mask, p_col_collision]
                    .dropna()
                    .astype(str)
                    .str.strip()
                    .tolist()
                )
        # Store non-driver ages for this year
        if non_driver_age_rows:
            all_non_driver_ages.append(pd.concat(non_driver_age_rows, ignore_index=True))
        # Store yearly bicyclist counts
        yearly_bike_counts.append(
            {
                "Year": year,
                "BicyclistPartyCount": bike_party_count,
                "UniqueBikeCrashCount": len(bike_collision_ids),
                "SFCollisionCount": len(sf_collision_ids),
                "SFPartyRowCount": sf_party_row_count,
            }
        )
        
        print(f"  Bicyclist parties: {bike_party_count:,}")
        print(f"  Unique bike crashes: {len(bike_collision_ids):,}")
        # Look up the exact locations (street names, coordinates, etc.)
        # for the bicycle-involved crashes found above.
        if bike_collision_ids:
            location_cols = [
                c
                for c in [
                    col_collision,
                    col_primary,
                    col_secondary,
                    col_secdir,
                    col_secdist,
                    col_lat,
                    col_lon,
                    col_dt,
                ]
                if c is not None
            ]
            bike_locations = sf_crashes[
                sf_crashes[col_collision].astype(str).str.strip().isin(bike_collision_ids)
            ][location_cols].copy()
            bike_locations["Year"] = year
            # Standardize column names
            rename_map = {
                col_collision: "CollisionId",
                col_primary: "PrimaryRoad",
                col_secondary: "SecondaryRoad",
                col_secdir: "SecondaryDirection",
                col_secdist: "SecondaryDistance",
                col_lat: "Latitude",
                col_lon: "Longitude",
                col_dt: "CrashDateTime",
            }
            bike_locations = bike_locations.rename(
                columns={k: v for k, v in rename_map.items() if k in bike_locations.columns}
            )
            # Build addresses
            bike_locations["Address"] = bike_locations.apply(build_address, axis=1)
            keep_cols = [
                c
                for c in [
                    "Year",
                    "CollisionId",
                    "Address",
                    "PrimaryRoad",
                    "SecondaryRoad",
                    "Latitude",
                    "Longitude",
                    "CrashDateTime",
                ]
                if c in bike_locations.columns
            ]
            bike_location_frames.append(
                bike_locations[keep_cols].drop_duplicates(subset=["CollisionId"])
            )
    # Step 1: Age frequency table
    if all_non_driver_ages:
        print("\n Processing age frequency data...")
        ages = pd.concat(all_non_driver_ages, ignore_index=True)
        ages["StatedAge"] = ages["StatedAge"].astype(str).str.strip()
        ages.loc[
            ages["StatedAge"].isin(["", "nan", "None"]), "StatedAge"
        ] = "(blank)"
        age_frequency = (
            ages.groupby("StatedAge", dropna=False)
            .size()
            .reset_index(name="Frequency")
        )
        # Clean up decimal formatting
        age_frequency["StatedAge"] = age_frequency["StatedAge"].apply(
            lambda v: str(v)[:-2]
            if str(v).endswith(".0") and str(v)[:-2].replace("-", "").isdigit()
            else str(v)
        )
        age_frequency = (
            age_frequency.groupby("StatedAge", as_index=False)["Frequency"]
            .sum()
            .reset_index(drop=True)
        )
        # Sort by numeric age where possible
        age_frequency["_age_num"] = pd.to_numeric(
            age_frequency["StatedAge"], errors="coerce"
        )
        age_frequency = (
            age_frequency.sort_values(
                ["_age_num", "StatedAge"], na_position="last"
            )
            .drop(columns=["_age_num"])
            .reset_index(drop=True)
        )
        nonblank_age_frequency = age_frequency[
            ~age_frequency["StatedAge"].isin(["(blank)", "", "nan", "None"])
        ].copy()
    else:
        age_frequency = pd.DataFrame(columns=["StatedAge", "Frequency"])
    # Step 2: Yearly bike counts
        nonblank_age_frequency = age_frequency.copy()
    bike_counts = pd.DataFrame(yearly_bike_counts).sort_values(
        "Year"
    ).reset_index(drop=True)
    # Step 3: Combine all bike locations
    if bike_location_frames:
        bike_locations_final = pd.concat(bike_location_frames, ignore_index=True)
        bike_locations_final = bike_locations_final.sort_values(
            ["Year", "CollisionId"]
        ).reset_index(drop=True)
    else:
        bike_locations_final = pd.DataFrame(
            columns=[
                "Year",
                "CollisionId",
                "Address",
                "PrimaryRoad",
                "SecondaryRoad",
                "Latitude",
                "Longitude",
                "CrashDateTime",
            ]
        )
    print(" Analysis complete")
    # --- Save pedestrian summary CSVs for visualization ---
    # outputs are written to disk
    OUT_DIR = OUTPUT_DIR
    OUT_DIR.mkdir(exist_ok=True)

    # Create overall and year-level summary versions compatible with viz section
    overall_summary = nonblank_age_frequency.copy()
    overall_path = OUT_DIR / "pedestrian_crash_summary_table.csv"
    overall_summary.to_csv(overall_path, index=False)

    yearly_summary_data = []
    for y in YEARS:
        subset = bike_counts[bike_counts["Year"] == y].copy()
        subset["source_year"] = y
        subset["age_group"] = "(not computed)"
        subset["sex"] = "(not computed)"
        subset["severity"] = "(not computed)"
        subset["crash_count"] = subset["BicyclistPartyCount"]
        yearly_summary_data.append(subset[["source_year", "age_group", "sex", "severity", "crash_count"]])
    summary_by_year = pd.concat(yearly_summary_data, ignore_index=True)
    yearly_path = OUT_DIR / "pedestrian_crash_summary_table_by_year.csv"
    summary_by_year.to_csv(yearly_path, index=False)

    print("\n Saved summaries:")

    return age_frequency, nonblank_age_frequency, bike_counts, bike_locations_final
# Run the analysis
age_frequency, nonblank_age_frequency, bike_counts, bike_locations = run_analysis()
```

## Cell 5: Creating Summary

**How Claude helped:** Claude explained the overall structure of this function so I could understand it as a complete pipeline rather than just a block of code. The function loops through each year, filters the statewide crashes dataset down to only San Francisco, collects the unique collision IDs for SF crashes, and then processes the much larger parties file in chunks — matching each party row back to an SF collision ID. Within that loop it tracks non-driver ages, counts bicyclist parties, and records which collision IDs involved bicycles. At the end it builds three outputs: an age frequency table, a yearly bicycle count summary, and a location table for all bike crashes.

Claude also pointed out that this function saves two summary CSV files at the end (`pedestrian_crash_summary_table.csv` and `pedestrian_crash_summary_table_by_year.csv`) that the visualization section depends on — so `run_analysis()` must be run before any charts can be generated.

**What I learned:** Large analysis functions are easier to understand when you read them as a pipeline: what goes in, what gets filtered, what gets counted, and what gets saved. I also learned why chunk-based processing matters — the parties files have nearly 500,000 rows per year, so reading them all at once would be slow.

The run_analysis function pulls together the complete San Francisco crash analysis.
It reads in crash and party data files for several years, filters them down to
only events that happened within San Francisco, and calculates various statistics
about people involved in those crashes.
Specifically, it:
    - Summarizes ages of non-driver participants (for example, pedestrians and passengers)
    - Counts the number of bicycle-related crashes each year
    - Records where those bicycle crashes occurred (by coordinates and street names)
    - Saves all these results into files you can later visualize or explore

```python
import pandas as pd

def save_results_and_display_summary(
    age_frequency: pd.DataFrame,
    nonblank_age_frequency: pd.DataFrame,
    bike_counts: pd.DataFrame,
    bike_locations: pd.DataFrame
) -> None:
    """Save analysis results to CSV files and display summary."""
    
    # Define output file paths
    age_path = OUTPUT_DIR / "sf_non_driver_stated_age_frequency_2021_2025.csv"
    age_nonblank_path = OUTPUT_DIR / "sf_non_driver_stated_age_frequency_nonblank_2021_2025.csv"
    counts_path = OUTPUT_DIR / "sf_bicyclist_involved_counts_by_year_2021_2025.csv"
    bike_locations_path = OUTPUT_DIR / "sf_bicycle_crash_locations_2021_2025.csv"
    # Save to CSV files
    age_frequency.to_csv(age_path, index=False)
    nonblank_age_frequency.to_csv(age_nonblank_path, index=False)
    bike_counts.to_csv(counts_path, index=False)
    bike_locations.to_csv(bike_locations_path, index=False)
    print("Output files saved:")
    print(f"  • {age_path}")
    print(f"  • {age_nonblank_path}")
    print(f"  • {counts_path}")
    print(f"  • {bike_locations_path}")
    
    print(f"\n Summary Statistics:")
    print(f"  • Total non-driver age records: {age_frequency['Frequency'].sum():,}")
    print(f"  • Non-blank age records: {nonblank_age_frequency['Frequency'].sum():,}")
    print(f"  • Total bicycle crash locations: {len(bike_locations):,}")
    
    print(f"\n Yearly Bicyclist Involvement:")
    display(bike_counts)
    
    print(f"\n Top 10 Non-Driver Ages (Non-blank):")
    if not nonblank_age_frequency.empty:
        display(nonblank_age_frequency.head(10))
    else:
        print("No non-blank age data available")
    # Save results and display summary
save_results_and_display_summary(age_frequency, nonblank_age_frequency, bike_counts, bike_locations)
```

## Cell 10: Loading Summary Tables

**How Claude helped:** Claude explained that this cell reloads the summary CSVs from disk back into memory as DataFrames. Even though `run_analysis()` already has these in memory as variables, explicitly reloading from the saved files at the start of the visualization section is a deliberate choice — it means the visualization section can run independently without having to re-run the full analysis. **What I learned:** Separating the analysis section (which is slow and processes large files) from the visualization section (which is fast and just needs the summaries) is good notebook design. Saving to CSV and reloading is the bridge between the two. This is part of the broader principle of keeping analysis and visualization in distinct parts of a notebook — or even separate notebooks.

```python
# Configuration
AGE_ORDER = ["0-17", "18-24", "25-34", "35-44", "45-54", "55-64", "65+", "Unknown"]
SEX_ORDER = ["Female", "Male", "Non-binary", "Unknown"]
SEVERITY_ORDER = [
    "Fatal",
    "SuspectSerious",
    "SevereInactive",
    "SuspectMinor",
    "PossibleInjury",
    "ComplaintOfPainInactive",
    "OtherVisibleInactive",
    "Unknown",
]
def make_pivot_plot(df, title, chart_type="Count"):
    """Create either a count or percent stacked bar chart of crash severity mix."""
    if df.empty:
        fig, ax = plt.subplots(figsize=(10, 5))
        ax.axis("off")
        ax.text(0.5, 0.5, "No data for this selection", ha="center", va="center", fontsize=14)
        plt.show()
        return
    df["age_group"] = pd.Categorical(df["age_group"], categories=AGE_ORDER, ordered=True)
    df["sex"] = pd.Categorical(df["sex"], categories=SEX_ORDER, ordered=True)
    pivot = (
        df.pivot_table(
            index=["age_group", "sex"],
            columns="severity",
            values="crash_count",
            aggfunc="sum",
            fill_value=0,
        )
        .reindex(columns=[c for c in SEVERITY_ORDER if c in df["severity"].unique()])
        .sort_index()
    )
    # Normalize to percent if selected
    if chart_type.lower() == "percent":
        pivot = pivot.div(pivot.sum(axis=1).replace(0, np.nan), axis=0) * 100
        ylabel = "Percent of Crashes"
    else:
        ylabel = "Crash Count"
    pivot.plot(kind="bar", stacked=True, figsize=(14, 6), width=0.85, alpha=0.9)
    plt.title(title, fontsize=14)
    plt.xlabel("Age Group | Sex")
    plt.ylabel(ylabel)
    plt.xticks(rotation=45, ha="right")
    plt.legend(title="Severity", bbox_to_anchor=(1.02, 1), loc="upper left")
    plt.tight_layout()
    plt.show()
```

## Cell 13: Exporting Bicycle Crash Locations for Tableau

**How Claude helped:** Claude explained why this cell exists as a separate step rather than just using the `bike_locations` DataFrame directly. Tableau requires clean, flat data — specifically, it cannot plot points where latitude or longitude are missing. The `.dropna(subset=["Latitude", "Longitude"])` step filters those records out before saving. Claude pointed out that this is why the output has only 54 rows even though there are 2,197 total bicycle crash locations. Most records are missing coordinates, likely because the original data source didn't geocode all crashes.

**What I learned:** Data exported for visualization tools often needs different preparation than data used for analysis. Tableau specifically needs complete coordinate pairs to map points, so filtering for non-null lat/lon before export is necessary. Knowing that only 54 out of 2,197 bike crashes have coordinates is itself a useful finding. The map will only show a small fraction of the actual crash locations.

```python
OUTPUT_DIR.mkdir(exist_ok=True)
yearly_severity_frames = []
# helper function looks up any column whose normalized name contains a given token
def find_col(df, token):
    matches = [c for c in df.columns if token in c.replace(" ", "").replace("_", "").lower()]
    return matches[0] if matches else None
for year in YEARS:
    print(f"\n--- Processing {year} ---")
    crash_path = RAW_DIR / f"crashes_{year}.csv"
    injured_path = RAW_DIR / f"injured_witnesses_{year}.csv"
    if not crash_path.exists() or not injured_path.exists():
        continue
    # From crashes: we only need City Name and Collision Id
    crash_df = pd.read_csv(crash_path, dtype=str, low_memory=False)
    col_collision_crash = find_col(crash_df, "collisionid")
    col_city = find_col(crash_df, "cityname") or find_col(crash_df, "city")
    if not col_collision_crash or not col_city:
        raise ValueError(f"{crash_path.name}: couldn't detect CollisionID or City Name column")
    crashes = crash_df[[col_collision_crash, col_city]].rename(
        columns={col_collision_crash: "collision_id", col_city: "city_name"}
    ).copy()
    crashes["collision_id"] = crashes["collision_id"].astype(str).str.strip()
    crashes["city_name"] = crashes["city_name"].astype(str).str.strip().str.upper()
    # --- Load the injured witness file ---
    inj_df = pd.read_csv(injured_path, dtype=str, low_memory=False)
    col_collision_inj = find_col(inj_df, "collisionid")
    col_injury = find_col(inj_df, "extentofinjury")
    if not col_collision_inj or not col_injury:
        raise ValueError(f"{injured_path.name}: missing CollisionID or ExtentOfInjuryCode column")
    injured = inj_df[[col_collision_inj, col_injury]].rename(
        columns={col_collision_inj: "collision_id", col_injury: "extentofinjurycode"}
    ).copy()
    injured["collision_id"] = injured["collision_id"].astype(str).str.strip()
    injured["extentofinjurycode"] = injured["extentofinjurycode"].astype(str).str.strip()
    # Merge both datasets on Collision Id
    # This brings City Name into the injured_witness rows
    merged = injured.merge(
        crashes[["collision_id", "city_name"]],
        on="collision_id",
        how="left",
        validate="m:1"  # many injuries can connect to one crash record
    )
    # Filter for San Francisco only
    sf_injuries = merged[merged["city_name"] == "SAN FRANCISCO"].copy()
    print(f"  → Total injured‑person records: {len(injured):,}")
    print(f"  → San Francisco matched injuries: {len(sf_injuries):,}")
    if sf_injuries.empty:
        continue
    # Count injuries per severity level (ExtentOfInjuryCode)
    severity_counts = (
        sf_injuries.groupby("extentofinjurycode", dropna=False)
        .size()
        .reset_index(name="CrashCount")
        .rename(columns={"extentofinjurycode": "Severity"})
    )
    severity_counts["Year"] = year
    yearly_severity_frames.append(severity_counts)
# Combine all years together
if yearly_severity_frames:
    severity_summary = pd.concat(yearly_severity_frames, ignore_index=True)
else:
    severity_summary = pd.DataFrame(columns=["Year", "Severity", "CrashCount"])
# Clean up severity values
severity_summary["Severity"] = (
    severity_summary["Severity"]
    .replace({"": "Unknown", "nan": "Unknown", None: "Unknown"})
    .fillna("Unknown")
)
severity_summary["Year"] = severity_summary["Year"].astype(str)
# dd an "All Years" total
totals = (
    severity_summary.groupby("Severity", as_index=False)["CrashCount"]
    .sum()
)
totals["Year"] = "All Years"
severity_final = pd.concat([severity_summary, totals], ignore_index=True)
# Save CSV for Datawrapper
output_path = OUTPUT_DIR / "sf_injury_severity_linked_summary_for_datawrapper.csv"
severity_final.to_csv(output_path, index=False)
print(f"File saved to: {output_path}")
print(f"Total rows: {len(severity_final):,}")
severity_final.head(10)
```

## Cell 14: Injury Severity Analysis — Linking Crashes to Injured Witnesses

**How Claude helped:** Claude walked me through why this cell needs to merge two separate files — crashes and injured witnesses — rather than just reading one. The `injured_witnesses` file has each injured person's severity code but doesn't include city name. The `crashes` file has city name but not individual injury details. To get SF-only injury severity counts, you have to join them on `CollisionId`. Claude explained the `validate="m:1"` argument in `.merge()`, which tells pandas to enforce that each injury row can only match to one crash row (many-to-one). This catches data quality issues if a collision ID somehow maps to multiple crash records.

**What I learned:** Real-world datasets are often normalized across multiple files, and analysis requires joining them on a shared key. Understanding which file holds which information is essential before you can write a correct join. The `validate` parameter in `.merge()` is a useful  check I hadn't used before.

```python
severity_path = OUTPUT_DIR / "sf_injury_severity_linked_summary_for_datawrapper.csv"
df = pd.read_csv(severity_path, dtype=str)
# Ensure Crash Count is numeric
df["CrashCount"] = pd.to_numeric(df["CrashCount"], errors="coerce").fillna(0).astype(int)

# Define final grouped severity labels
severity_map = {
    "Fatal": "Fatal",
    "SuspectSerious": "Severe",
    "SevereInactive": "Severe",
    "OtherVisibleInactive": "Severe",
    "SuspectMinor": "Minor",
    "PossibleInjury": "Minor",
    "ComplaintOfPainInactive": "Minor",
    "Unknown": "Unknown"
}
df["SeverityGroup"] = df["Severity"].map(severity_map).fillna("Unknown")

# Aggregate by Year + SeverityGroup
grouped = (
    df.groupby(["Year", "SeverityGroup"], as_index=False)["CrashCount"]
    .sum()
    .sort_values(["Year", "SeverityGroup"])
)
# Save grouped version for Datawrapper
grouped_path = OUTPUT_DIR / "sf_injury_severity_grouped_for_datawrapper.csv"
grouped.to_csv(grouped_path, index=False)
print(grouped_path)
grouped.head(20)
```
